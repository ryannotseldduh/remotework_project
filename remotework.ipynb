```python
import pandas as pd
import pprint
```

Reading in my csv's. I was able to get a lot of the cleaning done in this initial reading in. 

```python
df_2019 = pd.read_csv('CSV_data/2019_data.csv').drop(columns= ['Other','Other.1','Other.2','Other.3','Other.4','Other.5',"Anything else you want to add that we didn’t ask about?"]).dropna().replace(1,'Yes').replace(0,'No')
df_2020 = pd.read_csv('CSV_data/2020_data.csv').drop(columns= ['Other','Other.1','Other.2','Other.3','Other.4','Other.5',"Anything else you want to add that we didn’t ask about?"]).dropna().replace(1,'Yes').replace(0,'No')
df_2021 = pd.read_csv('CSV_data/2021_data.csv').drop(columns= ['Please link back to the original report when using this data: \nbuffer.com/2021-state-of-remote-work']).replace('1','Yes').replace('0','No')
df_2022 = pd.read_csv('CSV_data/2022_data.csv').drop(columns= ['Other - Write In:If you work from home, what room do you typically work in?','Prefer to self-describe:How do you identify?']).fillna('')
df_2023 = pd.read_csv('CSV_data/2023_data.csv').drop(columns=['Why would you not want to work remotely for the rest of your career? ']).fillna('')
```

The below cell shows me all the unique responses in this dataframe. I want to make these all match in terms of capitalization. The discrepancies are causing duplicate rows, seperating 'Consumer products' from 'Consumer products' and the like 

```python
df_2019['What industry do you work in?'].unique()
```

```python
df_2019['What industry do you work in?'].replace(to_replace='Consumer products',value='Consumer Products', inplace=True)
```

```python
df_2019['What industry do you work in?'].unique()
```

```python
industry_rows_2020 = {'Nonprofit, Medical and Healthcare':'Non-profit',
                 "Marketing, Media and Publishing":"Media and Publishing",
                 "Government, Law and Legal Services":"Government",
                 "Education, E-commerce":"Education",
                 "Consumer products":"Consumer Products"}
df_2020['What industry do you work in?'].rename(index = industry_rows_2020, inplace=True)
df_2020['What industry do you work in?'].unique()
```

This code is needed to further clean this csv. It had the columns labeled "Unnamed:1" etc. This code makes the information in row 0 into columns names. 

```python
df_2021.columns = df_2021.iloc[0]
```

This code drops a futher column I wasn't able to in the initial reading in. It also uses spaces to fill spaces rather than NaN. 

```python
df_2021 = df_2021.drop('Other', axis=1).fillna('')
```

This code merges my csv's that go together more easily. 

```python
con_csv = pd.concat([df_2019,df_2020,df_2022,df_2023])
con_csv.fillna('')
```

Needed to merge the 2021 data, when I tried to concatonate, the result messed up my columns

```python
all_csv = pd.concat([con_csv, df_2021])
all_csv.fillna('')
```

```python
all_csv['What industry do you work in?'].unique()
```

```python
all_csv['What industry do you work in?'].replace(to_replace='Consumer products',value='Consumer Products', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='E-commerce',value='Education, E-commerce', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='Education',value='Education, E-commerce', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='Law and Legal Services',value='Government, Law and Legal Services', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='Government',value='Government, Law and Legal Services', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='Marketing',value='Marketing, Media and Publishing', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='Media and Publishing',value='Marketing, Media and Publishing', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='Non-profit',value='Nonprofit, Medical and Healthcare', inplace=True)
all_csv['What industry do you work in?'].replace(to_replace='Medical and Healthcare',value='Nonprofit, Medical and Healthcare', inplace=True)
```

```python
all_csv['What industry do you work in?'].unique()
```

```python
all_csv['How do you identify?'].replace(to_replace='Prefer to self-describe',value='Prefer not to answer', inplace=True)
all_csv['How do you identify?'].replace(to_replace='Prefer to self-describe:',value='Prefer not to answer', inplace=True)
```

```python
all_csv['How do you identify?'].unique()
```
